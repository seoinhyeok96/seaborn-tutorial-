하나 이상의 플롯을 만들어 faceGrid 쉽게 수행할 수 있습니다.
추가 변수의 영향을 표시하려면 플롯의 의미론적 역활중 하나에 변수를 할당하는 대신 시각화를  패싯 하는데 사용하고 즉 여러축을 만들고 각 축에 대한 데이터의 하위 집합을 그립니다.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',col='time')
```

위에 방식으로 두 변수의 영향을 표시 할 수 도 있습니다. 하나는 열에 대한 면 분할을 통해 다른 하나는 행에 대한 면 분할을 통해 표시 할 수 있습니다.
그리드에 더 많은 변수를 추가하기 시작하면 더 그림 크기를 줄일 수 있습니다.

```python
sns.relplot(data=fmri, kind="line",x="timepoint", y="signal", hue="subject",col="region", row="event", height=3,
    estimator=None
)
```

변수의 여러 수준에서 효과를 조사하려는 경우 해당 변수를 열에서 패싯한 다음 패싯을 행으로 "래핑"하는 것이 좋습니다.

변수의 여러 수준에서 효과를 조사하려는 경우 해당 변수를 열에서 패싯한 다음 패싯을 행으로 "래핑"하는 것이 좋습니다.

```python
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", 
    col="subject", col_wrap=5,
     aspect=.75, linewidth=2.5,
)
```

```python
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```

"격자" 플롯 또는 "작은 배수"라고도 하는 이러한 시각화는 시각적으로 전체 패턴과 이러한 패턴의 편차를 모두 쉽게 감지할 수 있는 형식으로 데이터를 표시하기 때문에 매우 효과적입니다. scatterplot()및 에서 제공하는 유연성을 활용해야 하지만 relplot()일반적으로 여러 개의 간단한 플롯이 하나의 복잡한 플롯보다 더 효과적이라는 점을 항상 염두에 두십시오.
