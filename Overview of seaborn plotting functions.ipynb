## seaborn 플로팅 함수의 개요

### seaborn은 상호 작용은 대부분 일련의 플로팅 기능을 통해 발생합니다. 자습서 이후 자에서는 각 기능을에서 제공하는 특정 기능을 살펴 봅니다. 이장에서는 접하게 될 다양한 종류의 함수를 개괄적으로 소개합니다.

# 유사한 작업에 대한 유사한 기능

## seabron 네임스페이스는 평평합니다. 모든 기능은 최상위 수준에서 액섹스 할 수 있습니다. 그러나 코드 자체는 서로 다른 수단을 통해 유사한 시각화 목표를 달성하는 함수 모듈로 계층 구조로 구성되어 있습니다. 대부분의 문서는 이러한 모듈을 중심으로 구성되어 있습니다. "관계형","분산형",및 "변주형"과 같은 이름을 접하게 될것입니다.

### 예를 들어 distributions 모듈은 데이터 포인트의 분포를 나다내는데 특화된 함수를 정의합니다. 여기에는 히스토그램과 같은 친숙한 방법이 포함됩니다.

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins,x="flipper_length_mm",hue="species",multiple="stack")
```

## 커널 밀도 추청과 같은 유사하지만 덜 친숙한 옵션과 함께:

```python
sns.kdeplot(data=penguins, x="flipper_length_mm",hue="species",multiple="stack")
```

### multiple="stack" 모듈내 함수는 많은 기본 코드를 공유하고 라이브러리의 다른 구성 요소에 없을 수 있는 유사한 기능을 제공합니다.
### 다양한 표현에는 보완적인 강점과 약점이 있는 경우가 많기 때문에 데이터 세트를 탐색할 때 다양한 시각적 표현 간 전환을 용이하게 하도록 설계 되었습니다.
