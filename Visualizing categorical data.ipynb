범주형 데이터와 관련된 관계를 시각화하는 여러 가지 방법이 있습니다. 

relplot()와 둘 중 하나 scatterplot()또는 사이의 관계와 유사하게 lineplot()이러한 플롯을 만드는 두 가지 방법이 있습니다. 

다양한 방식으로 범주형 데이터를 플로팅하기 위한 여러 축 수준 함수와 catplot()이에 대한 통합된 상위 수준 액세스를 제공하는 그림 수준 인터페이스가 있습니다.

```python
import seaborn as sns
sns.set_style("darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill")
```

# Categorical scatterplots

한 범주에 속하는 모든 포인트가 범주형 변수에 해당하는 축을 따라 동일한 위치에 놓이게 됩니다. stripplot()에서 기본 "종류"인 에서 사용하는 접근 방식은 catplot()소량의 무작위 "지터"를 사용하여 범주형 축의 점 위치를 조정하는 것입니다.

```python
sns.catplot(data=tips, x="day", y="total_bill");
```

```python
#이 jitter매개변수는 지터의 크기를 제어하거나 완전히 비활성화합니다.

sns.catplot(data=tips, x="day", y="total_bill", jitter=False)
```

접근 방식은 포인트가 겹치지 않도록 방지하는 알고리즘을 사용하여 범주형 축을 따라 포인트를 조정합니다. 상대적으로 작은 데이터 세트에 대해서만 잘 작동하지만 관찰 분포를 더 잘 표현할 수 있습니다. 이러한 종류의 플롯은 때때로 "beeswarm"이라고 합니다

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
# hue 추가

sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="swarm");
```

```python
#Categorical데이터 유형이 있는 경우 카테고리의 기본 순서를 설정할 수 있습니다. 범주형 축에 전달된 변수가 숫자로 보이면 수준이 정렬됩니다. 그러나 데이터는 여전히 범주형으로 취급되며 범주형 축의 서수 위치(구체적으로는 0, 1, …)에 그려집니다.

sns.catplot(data=tips.query("size != 3"), x="size", y="total_bill");
```

```python
#매개변수를 사용하여 플롯별로 순서를 제어할 수도 있습니다 order

sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"]);
```

```python
# 그러나 범주형 변수를 세로 축에 배치
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```

# Comparing distributions

* 데이터 세트의 크기가 커짐에 따라 범주형 산점도는 각 범주 내의 값 분포에 대해 제공할 수 있는 정보가 제한됩니다.
* 이런 일이 발생하면 범주 수준에서 쉽게 비교할 수 있는 방식으로 분포 정보를 요약하는 몇 가지 방법이 있습니다.



# Boxplots

boxplot(). 이러한 종류의 플롯은 극단값과 함께 분포의 3사분위수 값을 보여줍니다
상위 사분위수의 1.5 IQR 내에 있는 지점까지 확장되며 이 범위를 벗어나는 관측값은 독립적으로 표시됩니다. 이것은 boxplot의 각 값이 데이터의 실제 관측치에 해당함을 의미합니다.

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
```

```python
# 의미 체계 를 추가할 때 hue의미 체계 변수의 각 수준에 대한 상자가 범주 축을 따라 이동하므로 겹치지 않습니다.

sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box");
```

```python
# 이 동작을 "도징"이라고 하며 시맨틱 변수가 기본 범주 변수 내에 중첩되어 있다고 가정하기 때문에 기본적으로 켜져 있습니다. 그렇지 않은 경우 회피를 비활성화할 수 있습니다.

tips["weekend"] = tips["day"].isin(["Sat", "Sun"])

sns.catplot(data=tips, x="day", y="total_bill", hue="weekend",kind="box", dodge=False,);
```

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(data=diamonds.sort_values("color"),x="color", y="price", kind="boxen");
```

# Violinplots

다른 접근 방식은 분포violinplot() 자습서 에 설명된 커널 밀도 추정 절차와 boxplot을 결합니다

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="sex", kind="violin",);
```

이 접근법은 커널 밀도 추정치를 사용하여 값 분포에 대한 보다 풍부한 설명을 제공합니다. 또한 상자 그림의 사분위수 및 위스커 값이 바이올린 내부에 표시됩니다. 단점은 violinplot이 KDE를 사용하기 때문에 간단한 boxplot에 비해 약간의 복잡성을 추가하여 조정해야 할 수 있는 다른 매개 변수가 있다는 것입니다.

```python
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex",
    kind="violin", bw=.15, cut=0,);
```

### 또한 색상 매개변수에 두 가지 수준만 있는 경우 바이올린을 "분할"할 수 있어 공간을 보다 효율적으로 사용할 수 있습니다.



```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=True,)
```

```python
# 마지막으로 요약 상자 그림 값 대신 각 개별 관찰을 표시하는 방법을 포함하여 바이올린 내부에 그려지는 그림에 대한 몇 가지 옵션이 있습니다.

sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", inner="stick", split=True, palette="pastel",
);
```

분포 요약과 함께 각 관찰을 표시하기 위해 박스 플롯 또는 바이올린 플롯과 swarmplot()결합 하는 것도 유용할 수 있습니다 .stripplot()

```python
g = sns.catplot(data=tips, x="day", y="total_bill", kind="violin", inner=None)
sns.swarmplot(data=tips, x="day", y="total_bill", color="k", size=3, ax=g.ax);
```
