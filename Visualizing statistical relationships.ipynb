## 세가지 seaborn 함수에 대해 설명합니다. 우리가 가장 많이 사용할 것은 relplot(). 산점도와 선도의 두가지 일반적인 접근 방식을 사용하여 관계를 시각화하기 위한 수치 수준의 기능 입니다. relplot()를 FacetGrid다음 두 축 수준 함수 중 하나와 결합합니다.

scatterplot()은 kind="scatter" 기본값
lineplot()은 kind='line'이 기본값

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip")
```

포인트가 2차원으로 표시되는 동안 세 번째 변수에 따라 포인트에 색상을 지정하여 플롯에 다른 차원을 추가할 수 있습니다. seaborn에서는 포인트의 색상이 의미를 갖기 때문에 "색상 의미 체계"를 사용한다고 합니다.

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

클래스 간의 차이점을 강조하고 접근성을 개선하기 위해 각 클래스에 대해 서로 다른 마커 스타일을 사용할 수 있습니다.

```python
sns.relplot(data=tips,x="total_bill", y="tip", hue="smoker", style='smoker')
```

클래스 간의 차이점을 강조하고 접근성을 개선하기 위해 각 클래스에 대해 서로 다른 마커 스타일을 사용할 수 있습니다.

```python
sns.relplot(data=tips,x="total_bill", y="tip", hue="smoker", style="time")
```

위의 예에서 색조 시맨틱은 범주형이므로 기본 정 성적 팔레트가 적용되었습니다. 색조 시맨틱이 숫자인 경우(특히 부동으로 캐스팅할 수 있는 경우) 기본 색상이 순차적 팔레트로 전환됩니다.



```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="size")
```

두 경우 모두 색상 팔레트를 사용자 정의할 수 있습니다. 이를 위한 많은 옵션이 있습니다. 여기서 문자열 인터페이스를 사용하여 순차 팔레트를 다음과 같이 사용자 정의합니다 cubehelix_palette().

```python
sns.relplot(data=tips,x="total_bill", y="tip",hue="size", palette="ch:r=-.5,l=.75")
```

세 번째 종류의 의미 변수는 각 포인트의 크기를 변경합니다.

```python
sns.relplot(data=tips, x="total_bill", y="tip", size="size",hue="size")
```

와 달리 matplotlib.pyplot.scatter()변수의 리터럴 값은 점의 영역을 선택하는 데 사용되지 않습니다. 대신 데이터 단위의 값 범위가 면적 단위의 범위로 정규화됩니다. 이 범위는 사용자 정의할 수 있습니다.

```python
sns.relplot(data=tips, x="total_bill", y="tip",size="size", sizes=(15, 200))
```
