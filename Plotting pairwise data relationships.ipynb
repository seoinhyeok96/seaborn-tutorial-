PairGrid또한 동일한 플롯 유형을 사용하여 각각의 데이터를 시각화하는 작은 서브플롯의 그리드를 빠르게 그릴 수 있습니다. 에서 PairGrid각 행과 열은 다른 변수에 할당되므로 결과 플롯은 데이터 세트의 각 쌍별 관계를 보여줍니다. 이 스타일의 플롯은 각 관계를 표시하는 가장 일반적인 방법이지만 PairGrid산점도에 국한되지 않기 때문에 "산점도 매트릭스"라고도 합니다.

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style("darkgrid")
```

```python
iris = sns.load_dataset("iris")
g = sns.PairGrid(iris)
g.map(sns.scatterplot);
```

각 열에서 변수의 일변량 분포를 표시하기 위해 대각선에 다른 함수를 그릴 수 있습니다. 그러나 축 틱은 이 플롯의 개수 또는 밀도 축과 일치하지 않습니다.

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot);
```

```python
# 이 플롯을 사용하는 매우 일반적인 방법은 별도의 범주형 변수로 관측값에 색상을 지정합니다. 
#예를 들어 붓꽃 데이터셋에는 붓꽃의 3가지 종 각각에 대한 4가지 측정값이 있으므로 이들이 어떻게 다른지 확인할 수 있습니다.

g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend();
```

```python
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species")
g.map(sns.scatterplot);
```

```python
#관계의 다른 측면을 강조하기 위해 위쪽 삼각형과 아래쪽 삼각형에 다른 기능을 사용하는 것도 가능합니다.
g = sns.PairGrid(iris)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False);
```

```python
#대각선에 항등 관계가 있는 정사각형 격자는 실제로는 특별한 경우일 뿐이며 행과 열에 다른 변수를 사용하여 플롯할 수 있습니다.
tips = sns.load_dataset("tips")
g = sns.PairGrid(tips, y_vars=["tip"], x_vars=["total_bill", "size"], height=4)
g.map(sns.regplot, color=".3")
g.set(ylim=(-1, 11), yticks=[0, 5, 10]);
```

```python
#물론 미적 속성은 구성 가능합니다. 예를 들어 다른 팔레트를 사용하여(예를 들어 변수의 순서를 표시하기 위해 hue) 플로팅 함수에 키워드 인수를 전달할 수 있습니다.

g = sns.PairGrid(tips, hue="size", palette="GnBu_d")
g.map(plt.scatter, s=50, edgecolor="white")
g.add_legend();
```

```python
#PairGrid은 유연하지만 데이터 세트를 빠르게 살펴보기 위해 사용하기가 더 쉬울 수 있습니다 pairplot(). 이 기능은 기본적으로 산점도와 히스토그램을 사용하지만 몇 가지 다른 종류가 추가될 예정입니다(현재는 비대각선에 회귀 플롯을 그리고 대각선에 KDE를 그릴 수도 있습니다).

sns.pairplot(iris, hue="species", height=2.5);
```

```python
#또한 키워드 인수를 사용하여 플롯의 미학을 제어할 수 있으며 PairGrid추가 조정을 위해 인스턴스를 반환합니다.

g = sns.pairplot(iris, hue="species", palette="Set2", diag_kind="kde", height=2.5)
```
