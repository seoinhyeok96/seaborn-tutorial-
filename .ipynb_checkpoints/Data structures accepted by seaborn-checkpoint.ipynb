# Long-form vs. wide-form data

- seaborn의 대부분의 플로팅 기능은 데이터 벡터를 지향합니다.
- x에 대해 플로팅할 때 y각 변수는 백터여야 합니다.
- seaborn은 일부 표 형식으로 구성된 둘 이상의 백터가 있는 데이터 시트를 허용합니다.
- "긴 형식"과"넓은 형식" 데이터 테이블 사이에는 근본적인 차이가 있으며 seaborn은 각각을 다르게 취급합니다. 

# Long-form data
-  긴 형식의 데이터 테이블에는 다음과 같은 특징이 있습니다.

- 각 변수는 열입니다.
- 각 관찰은 행입니다.

```python
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

- 긴 형식의 데이터를 사용하면 테이블의 열의 변수 중 하나에 명시적으로 할당하여 플롯에서 역할이 부여됩니다.
- 예를들어, 연간 승객수 월별 도표를 작성하면 다음과 같습니다

```python
sns.relplot(data=flights,x='year',y='passengers',hue='month',kind='line');
```

# Wide-form data

- 간단한 데이터 세트의 경우 열과 행의 다양한 수준의 변수가 포함된 스프레드시트에서 데이터를 보는 방식으로 데이터를 생각하는 것이 더 직관적인 경우가 많습니다.
- 예를 들어 항공편 데이터 세트를'피벗'하여 넓은 형식의 조직으로 변환하여 각 열에 연도별 월별 시계열이 포함하도록 할 수 있습니다.

```python
flights_wide = flights.pivot(index='year',columns='month',values ='passengers')
flights_wide .head()
```

- 여기에 동일한 세 가지 변수가 있지만 다르게 구성되어 있습니다. 이 데이터 세트의 변수의 명명된 필드가 아닌 테이블의 차원에 연결됩니다.
- 각 관찰은 테이블의 셀에 있는 값과 행 및 열 인덱스에 대한 셀의 좌표로 정의됩니다.

- 긴 형식의 데이터를 사용하면 이름으로 데이터 세트의 변수에 액세스 할 수 있습니다. 와이드 형식 데이터의 경우에는 그렇지 않습니다.
- 그럼에도 불구하고 테이블의 차원과 데이터 세트 변수 간에 명확한 연관성이 있기 때문에 seaborn은 플롯에서 이러한 변수 역할을 할당할 수 있습니다.

```python
sns.relplot(data=flights_wide,kind='line');
```

- 이 플롯은 이전 플롯과 매우 유사해 보입니다. xSeaborn은 데이터 프레임의 인덱스를 에 데이터 프레임의 값을 에 할당했으며 y매월 별도의 선을 그렸습니다. 그러나 두 플롯 사이에는 눈에 띄는 차이가 있습니다. 데이터 세트가 긴 형식에서 넓은 형식으로 변환하는 "피벗" 작업을 거쳤을 때 값의 의미에 대한 정보가 손실되었습니다. 결과적으로 y축 레이블이 없습니다. (줄에는 대시도 있습니다. 플롯에 더 쉽게 액세스할 수 있도록 relplot()열 변수를 the hue및 stylesemantic 모두에 매핑했기 때문입니다. 긴 형식의 경우에는 그렇게 하지 않았지만 를 설정하여 가질 수 있습니다. style="month")

- 지금까지 우리는 넓은 형식의 데이터를 사용하면서 타이핑을 훨씬 적게 했고 거의 같은 플롯을 만들었습니다. 이것은 더 쉬운 것 같습니다! 그러나 긴 형식 데이터의 큰 장점은 데이터를 올바른 형식으로 가지고 있으면 더 이상 구조에 대해 생각할 필요가 없다는 것 입니다 . 플롯에 포함된 변수에 대해서만 생각하여 플롯을 설계할 수 있습니다. 예를 들어, 각 연도의 월별 시계열을 나타내는 선을 그리려면 간단히 변수를 재할당하십시오.

```python
sns.relplot(data=flights,x='month',y='passengers',hue='year',kind='line');
```

- 와이드 형식 데이터세트로 동일한 리매핑을 달성하려면 테이블을 바꿔야 합니다.

```python
sns.relplot(data=flights_wide.transpose(),kind='line');
```

(이 예는 또한 또 다른 주름을 보여줍니다. 즉, seaborn은 현재 데이터 유형에 관계없이 넓은 형식 데이터 세트의 열 변수를 범주형으로 간주하는 반면, 긴 형식 변수는 숫자이기 때문에 양적 색상 팔레트가 할당되고 범례. 향후 변경될 수 있음).



명시적 변수 할당이 없다는 것은 각 플롯 유형이 와이드 형식 데이터의 차원과 플롯의 역할 간에 고정 매핑을 정의해야 함을 의미합니다. 이 자연스러운 매핑은 플롯 유형에 따라 다를 수 있으므로 넓은 형식의 데이터를 사용할 때 결과를 예측하기 어렵습니다. 예를 들어 범주형 도표는 테이블의 열 차원을 할당한 x다음 행 전체에서 집계합니다(인덱스 무시).

```python
sns.catplot(data = flights_wide,kind='box');
```

넓은 형식의 데이터를 표현하기 위해 pandas를 사용하는 경우 몇 가지 변수(3개 이하)로 제한합니다. 이는 seaborndl pandas가 테이블 형식으로 추가 변수를 나타내는 방식인 다중 인덱스 정보를 사용하지 않기 때문입니다.현재 seaborn은 의 객체를 직접 지원하지 않지만 메서드를 사용하여 xarraylong-form으로 변환한 다음 다른 long-form 데이터 세트와 마찬가지로 seaborn에서 플롯할 수 있습니다.pandas.DataFrameto_pandas
