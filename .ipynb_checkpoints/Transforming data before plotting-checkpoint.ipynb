## Transforming data before plotting
- 통계 변환
- 많은 seaborn 함수와 마찬가지로 객체 인터페이스는 통계적 변환을 지원합니다. Stat이들은 다음과 같은 개체 에 의해 수행됩니다 Agg.

```python
import seaborn as sns
import seaborn.objects as so
penguins = sns.load_dataset("penguins")
tips = sns.load_dataset('tips')
```

```python
so.Plot(penguins, x='species',y='body_mass_g').add(so.Bar(),so.Agg())
```

그럼에도 불구하고 다음 과 같이 구현된 stat및 두번째 변환을 사용하여 Bar 마크를 구성 할 수 있습니다. AggDodge

```python
so.Plot(penguins, x='species',y='body_mass_g').add(so.Dot(pointsize=10),so.Agg())
```

```python
so.Plot(penguins, x='species',y='body_mass_g',color='sex').add(so.Dot(pointsize=10),so.Agg())
```

# Resolving overplotting

일부 seaborn 기능에는 seaborn.barplot()막대를 한 번 "dodges" hue할 때와 같이 오버 플로팅을 자동으로 해결하는 메커니즘이 있습니다. 개체 인터페이스에는 덜 복잡한 기본 동작이 있습니다. 여러 그룹을 나타내는 막대는 기본적으로 겹칩니다.

```python
so.Plot(penguins, x='species',y='body_mass_g',color='sex').add(so.Bar(),so.Agg())
```

그럼에도 불구하고 다음 과 같이 구현된 stat 및 두 번째 변환을 사용하여 Bar마크를 구성할 수 있습니다 .AggDodge

```python
so.Plot(penguins, x="species", y="body_mass_g", color="sex").add(so.Bar(), so.Agg(), so.Dodge())
```

클래스 는 a와 비슷 하지만 조정 및 조정만 수행하는 변환 Dodge의 예입니다 . 클래스 는 모든 마크와 함께 적용될 수 있으며 첫 번째를 사용할 필요가 없습니다 .MoveStatxyMoveStat


```python
so.Plot(penguins, x="species", y="body_mass_g", color="sex").add(so.Dot(), so.Dodge())
```

Move여러 작업을 순서대로 적용하는 것도 가능합니다 .

```python
so.Plot(penguins, x="species", y="body_mass_g", color="sex").add(so.Dot(), so.Dodge(), so.Jitter(.3))
```

# Creating variables through transformation

통계 에는 와 가 Agg모두 정의되어 있어야 하지만 통계 변환을 통해 변수를 만들 수도 있습니다 . 예를 들어 통계는 또는 중 하나만 정의하면 되며 관찰을 세어 다른 하나를 생성합니다.xyHistx y

```python
so.Plot(penguins, x="species").add(so.Bar(), so.Hist())
```

통계 는 또한 숫자 데이터가 주어지면 새 값(비닝을 통해)을 Hist생성합니다 .

```python
so.Plot(penguins, x="flipper_length_mm").add(so.Bars(), so.Hist())
```

연속 축이 있는 플롯 Bars대신 를 어떻게 사용했는지 확인하십시오 . 이 두 마크는 관련이 있지만 기본값이 다르며 연속 히스토그램에 더 잘 작동합니다. 또한 보다 효율적인 다른 matplotlib 아티스트를 생성합니다. 다른 곳에서 단수/복수 표시의 패턴을 찾을 수 있습니다. 복수 버전은 일반적으로 마크 수가 많은 경우에 최적화되어 있습니다.BarxBars

일부 변환은 x및 를 모두 허용 y하지만 각 좌표에 대한 간격 데이터를 추가합니다. 이는 집계 후 오류 막대를 플로팅하는 데 특히 적합합니다.

```python
so.Plot(penguins, x="body_mass_g", y="species", color="sex").add(so.Range(), so.Est(errorbar="sd"), so.Dodge()).add(so.Dot(), so.Agg(), so.Dodge())
```

# Orienting marks and transforms
- 막대를 집계, 회피 및 그릴 때 x및 y변수는 다르게 처리됩니다. 각 작업에는 오리엔테이션 개념이 있습니다 . 는 Plot변수의 데이터 유형에 따라 방향을 자동으로 결정하려고 시도합니다. species예를 들어 and 의 할당을 뒤집으면 body_mass_g동일한 플롯을 얻을 수 있지만 수평 방향입니다.

```python
so.Plot(penguins, x="body_mass_g", y="species", color="sex").add(so.Bar(), so.Agg(), so.Dodge())
```

x경우에 따라 및 변수가 모두 숫자인 경우와 같이 올바른 방향이 모호합니다 y. 이러한 경우 orient매개변수를 다음에 전달하여 명시적으로 지정할 수 있습니다 Plot.add()

```python
so.Plot(tips, x="total_bill", y="size", color="time").add(so.Bar(), so.Agg(), so.Dodge(), orient="y")
```
