* 다른 응용 프로그램의 경우 각 범주 내의 분포를 표시하는 대신 값의 중심 경향에 대한 추정치를 표시할 수 있습니다. Seaborn에는 이 정보를 표시하는 두 가지 주요 방법이 있습니다. 중요한 것은 이러한 기능에 대한 기본 API가 위에서 설명한 것과 동일하다는 것입니다.

# Bar plots

이 목표를 달성하는 친숙한 스타일의 플롯은 막대 플롯입니다. seaborn에서 barplot()함수는 전체 데이터 세트에서 작동하고 함수를 적용하여 추정치를 얻습니다(기본적으로 평균을 취함). 각 범주에 여러 관측치가 있는 경우 부트스트래핑을 사용하여 추정치 주변의 신뢰 구간을 계산하고 오차 막대를 사용하여 플롯합니다.

```python
import seaborn as sns
sns.set_style("darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar");
```

```python
# 기본 오차 막대는 95% 신뢰 구간을 표시하지만 직접 설정 할 수 있다 

sns.catplot(data=titanic, x="age", y="deck", errorbar=("pi", 95), kind="bar");# pi 오차
```

```python
# 막대 그래프의 특별한 경우는 두 번째 변수에 대한 통계를 계산하는 대신 각 범주의 관측치 수를 표시하려는 경우입니다. 

sns.catplot(data=titanic, x="deck", kind="count", palette="ch:.25");
```

```python
sns.catplot(
    data=titanic, y="deck", hue="class", kind="count",
    palette="pastel", edgecolor=".6",);
```

# Point plots

```python
#동일한 정보를 시각화하기 위한 대체 스타일이 함수에서 제공됩니다 pointplot(). 
#이 함수는 또한 다른 축에 높이가 있는 추정치 값을 인코딩하지만 전체 막대를 표시하는 대신 점 추정치와 신뢰 구간을 표시합니다. 

sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="point");
```

```python
# 범주형 함수는 관계형 함수의 의미 체계가 부족하지만 style색상과 함께 마커 및/또는 선 스타일을 변경하여 최대로 액세스할 수 있다

sns.catplot(
    data=titanic, x="class", y="survived", hue="sex",
    palette={"male": "g", "female": "m"},
    markers=["^", "o"], linestyles=["-", "--"],
    kind="point")
```

# Showing additional dimensions

```python
tips = sns.load_dataset("tips")
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="smoker",
    kind="swarm", col="time", aspect=.7,);
```

```python
# 플롯을 추가로 사용자 지정하려면 FacetGrid플롯이 반환하는 객체에 대한 메서드를 사용할 수 있습니다.

g = sns.catplot(
    data=titanic,
    x="fare", y="embark_town", row="class",
    kind="box", orient="h",
    sharex=False, margin_titles=True,
    height=1.5, aspect=4,)

g.set(xlabel="Fare", ylabel="")
g.set_titles(row_template="{row_name} class")
for ax in g.axes.flat:
    ax.xaxis.set_major_formatter('${x:.0f}')
```
