### 데이터를 분석하거나 모델링하려는 노력의 초기 단계는 변수가 어떻게 분포되어 있는지 이해하는 것입니다. 분포 시각화 기술은 많은 중요한 질문에 대한빠른 답변을 제공할 수 있습니다.

# Plotting univariate histograms

분포를 시각화하는건 가장 일반적인 것으 히스토그램이다. displot은 동일한 기본 코드를 사용하는 기본 접근 방식입니다. histplot은 히스토그램은 데이터의 변수를 나타내는 축이 불연속 빈 세트로 나뉘고 각 빈에 속하는 관측값 수가 해당 막대 높이를 사용하여 표시되는 막대 그래프입니다.

```python
# import seaborn
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.displot(penguins,x='flipper_length_mm')
```

## Choosing the bin size (빈 크기 선택)

```python
sns.displot(penguins, x="flipper_length_mm", binwidth=3)
```

다른 상황에서는 bin의 크기보다 bin의 수를 지정하는 것이 더 합리적일 수 도 있습니다.

```python
sns.displot(penguins, x="flipper_length_mm", bins=20)
```

기본값이 실패하는 상황의 한 가지 예는 변수가 상대적으로 적은 수의 정수 값을 사용하는 경우입니다. 이 경우 기본 빈 너비가 너무 작아 분포에 어색한 간격이 생길 수 있습니다.

```python
tips = sns.load_dataset("tips")
sns.displot(tips, x="size")
```

한 가지 접근 방식은 배열을 다음에 전달하여 정확한 빈 나누기를 지정하는 것입니다 bins.

```python
sns.displot(tips, x="size", bins=[1, 2, 3, 4, 5, 6, 7])
```

discrete=True이는 해당 값을 중심으로 막대가 있는 데이터 세트의 고유한 값을 나타내는 빈 나누기를 선택하는 를 설정하여 수행할 수도 있습니다 .

```python
sns.displot(tips, x="size", discrete=True)
```

범주형 특성을 강조하기 위해 막대를 약간 "축소"하는 것도 도움이 될 수 있습니다.

```python
sns.displot(tips, x="day", shrink=.7)
```
