변수를 할당하면 각각의 고유한 값에 대해 별도의 히스토그램을 그리고 색상으로 구분합니다.histplot()hue

```python
# import seaborn
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.displot(penguins, x='flipper_length_mm',hue='species')
```

기본적으로 서로 다른 히스토그램은 서로 "계층화"되며 경우에 따라 구별하기 어려울 수 있습니다. 한 가지 옵션은 히스토그램의 시각적 표현을 막대 그래프에서 "단계" 플롯으로 변경하는 것입니다.

```python
sns.displot(penguins, x='flipper_length_mm',hue='species',element='step')
```

또는 각 막대를 레이어링하는 대신 "스택"하거나 수직으로 이동할 수 있습니다. 이 플롯에서 전체 히스토그램의 윤곽선은 단일 변수만 있는 플롯과 일치합니다.

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

누적 히스토그램은 변수 간의 부분-전체 관계를 강조하지만 다른 기능을 가릴 수 있습니다

또 다른 옵션은 막대를 "닷지"하여 수평으로 이동하고 줄입니다. 이렇게 하면 겹치는 부분이 없고 높이 측면에서 막대를 비교할 수 있습니다. 그러나 범주형 변수의 수준 수가 적은 경우에만 잘 작동합니다.

```python
sns.displot(penguins, x="flipper_length_mm", hue="sex",multiple="dodge")
```

두 번째 변수를 또는 에 할당하여 FacetGrid별도의 서브플롯에 각 개별 분포를 그릴 수도 있습니다 . 

```python
sns.displot(penguins, x="flipper_length_mm", col="sex")
```

```python
sns.displot(penguins, x="flipper_length_mm", col="sex",hue="sex")
```

# Normalized histogram statistics

하위 집합의 관측수가 같지 않을 경우 개수 측면에서 분포를 비교하는 것이 이상적이지 않을 수 있다는 것입니다. 한가지 해결책은 매개변수를 사용하여 카운트를 정규화 하는것입니다.

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density")
```

### 정규화 하면 전체 분포에 적용되므로 단순히 막대의 높이를 다시 조절합니다. 이때 common_norm=False 이용하면 하위집합이 독립적으로 정규화됩니다.

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density",common_norm=False)
```

## 밀도 정규화는 영역의 합이 1이 되도록 막대의 크기를 조정합니다. 결과적으로 밀도 축을 해석할 수 없습니다. 또 다른 옵션은 높의 합이 1이 되도독 막대를 정규화 하는 것입니다.
## 이것은 변수가 이산형일 때 의미가 있지만 모든 히스토그램에 대한 옵션입니다.

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="probability")
```
