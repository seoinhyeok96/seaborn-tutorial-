histplot()seaborn의 다른 여러 그림 수준 플로팅 함수는 및 함수를 사용합니다 kdeplot().

```python
import seaborn as sns
sns.set_style("darkgrid")
```

# Plotting joint and marginal distributions

- jointplot()은 두변수의 한계분포로 이변량 관계 또는 분포도를 증가시키는 것입니다.
- jointplot()이변량 분포 scatterplot()와 주변 분포를 나타냅니다 histplot().

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

### 유사하게 displot(), 다른 kind="kde"값을 설정하면 jointplot()조인트 및 마지널 플롯 모두 사용이 변경됩니다 kdeplot().

```python
sns.jointplot(data=penguins,x="bill_length_mm", y="bill_depth_mm", hue="species",kind="kde")
```

- jointplot()JointGrid직접 사용할 때 더 많은 유연성을 제공하는 클래스 에 대한 편리한 인터페이스입니다 .

```python
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

한계 분포를 표시하는 덜 눈에 띄는 방법은 각 개별 관찰을 나타내기 위해 플롯의 가장자리에 작은 눈금을 추가하는 "러그" 플롯을 사용합니다. 이것은 다음에 내장되어 있습니다 displot().

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm",kind="kde", rug=True)
```

###  그리고 축 수준 rugplot()함수를 사용하여 다른 종류의 플롯 측면에 깔개를 추가할 수 있습니다.

```python
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
sns.rugplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

# Plotting many distributions

- pairplot()함수는 결합 및 주변 분포의 유사한 혼합을 제공합니다. 그러나 단일 관계에 초점을 맞추는 대신 pairplot()"작은 다중" 접근 방식을 사용하여 모든 쌍별 관계와 함께 데이터 세트에 있는 모든 변수의 단변량 분포를 시각화합니다.

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
