```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style("darkgrid")
```

위의 도표는 한 쌍의 변수 사이의 관계를 탐색하는 다양한 방법을 보여줍니다. 그러나 종종 더 흥미로운 질문은 "이 두 변수 사이의 관계가 세 번째 변수의 함수로 어떻게 변하는가?"입니다. 여기에서 regplot()와 사이의 주요 차이점이 lmplot()나타납니다. regplot()항상 단일 관계를 표시 하지만 와 lmplot()결합하여 매핑 또는 패싯을 사용하여 여러 맞춤을 표시합니다

```python
tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

달리 relplot()산점도의 스타일 속성에 고유한 변수를 매핑할 수는 없지만 hue표식 모양으로 변수를 중복 코딩할 수 있습니다.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,markers=["o", "x"], palette="Set1");
```

다른 변수를 추가하기 위해 그리드의 행 또는 열에 나타나는 변수의 각 수준으로 여러 "패싯"을 그릴 수 있습니다.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time", row="sex", data=tips, height=4);
```

# 다른 맥락에서 회귀를 플로팅하기

몇 가지 다른 Seaborn 함수는 regplot()더 크고 복잡한 플롯의 맥락에서 사용됩니다. 첫 번째는 배포 자습서jointplot() 에서 소개한 함수 입니다 . 앞에서 설명한 플롯 스타일 외에도 다음을 전달하여 관절 축에 대한 선형 회귀 적합성을 표시하는 데 사용할 수 있습니다 .jointplot()regplot()kind="reg"

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg");
```

결합 및 와 pairplot()함께 함수를 사용하여 데이터 세트에서 변수 간의 선형 관계를 표시합니다. 이것이 와 어떻게 다른지 주의하십시오 . 아래 그림에서 두 축은 세 번째 변수의 두 수준에서 동일한 관계를 보여주지 않습니다. 오히려 데이터 세트에서 변수의 서로 다른 쌍 사이의 여러 관계를 표시하는 데 사용됩니다.kind="reg"regplot()PairGridlmplot()PairGrid()

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],height=5, aspect=.8, kind="reg");
```

추가 범주형 변수에 대한 조건화는 다음 hue매개변수를 사용하여 이러한 두 함수에 내장됩니다.

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],hue="smoker", height=5, aspect=.8, kind="reg");
```
