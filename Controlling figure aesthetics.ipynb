매력적인 인물을 그리는 것이 중요합니다. 자신을 위해 그림을 만들 때 데이터 세트를 탐색할 때 보기 좋은 플롯이 있는 것이 좋습니다. 시각화는 또한 정량적 통찰력을 청중에게 전달하는 데 핵심적이며, 이러한 설정에서는 관심을 끌고 뷰어를 끌어들이는 수치가 훨씬 더 필요합니다.

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style("darkgrid")
```

조정할 수 있는 다양한 스타일 매개변수를 확인하는 데 도움이 되는 일부 오프셋 사인파를 플로팅하는 간단한 함수를 정의해 보겠습니다.

```python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)
```

```python
sinplot()
```

```python
sinplot()
sns.despine()
```

# Temporarily setting figure style

앞뒤로 쉽게 전환할 수 있지만 명령문 axes_style()에서 함수를 사용하여 with플롯 매개변수를 임시로 설정할 수도 있습니다. 이렇게 하면 스타일이 다른 축으로 도형을 만들 수도 있습니다.

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])
    sinplot(6)

with sns.axes_style("white"):
    ax = f.add_subplot(gs[0, 1])
    sinplot(6)

with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[1, 0])
    sinplot(6)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(gs[1, 1])
    sinplot(6)
    
f.tight_layout()    
    
    
    
    
```

# Overriding elements of the seaborn styles


seaborn 스타일을 사용자 지정하려는 경우 및 rc인수에 매개 변수 사전을 전달할 수 있습니다 . 이 방법을 통해 스타일 정의의 일부인 매개변수만 재정의할 수 있습니다. (그러나 상위 수준 함수는 모든 matplotlib 매개변수의 사전을 사용합니다.)axes_style()set_style()set_theme()

포함된 매개변수를 확인하려면 인수 없이 함수를 호출하면 됩니다. 그러면 현재 설정이 반환됩니다.

```python
# 현재 설정확인
sns.axes_style()
```

```python
# 매개변수의 다른 버전을 설정

sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sinplot()
```
