# Figure-level vs. axes-level functions

### 다양한 모듈 외에도  seaborn 기능의 교차 분류가 '축 수준' 또는 '그림 수준'으로 분류됩니다.. 위의 예는 축수준 함수 입니다.
###  matplotlib.pyplot.Axes 함수의 반환 값인 단일 개체에 데이터를 표시합니다.

### 예를 들어, dosplot()분포 모듈의 그림 수준 함수 입니다. histplot() 기본 동작은 뒤에서 와 동일한 코드를 사용하여 히스토그램을 그리는 것입니다.

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
```

```python
sns.displot(data = penguins, x="flipper_length_mm", hue="species", multiple="stack" )
```

### 대신 커널 밀도 플롯을 그리려면 와 동일한 코드를 사용하고 다음 매개 변수인 kdeplot()를 사용해야 합니다.(kind)

```python
sns.displot(data = penguins, x="flipper_length_mm", hue="species", multiple="stack",kind = 'kde')
```

### Figure 수준 플롯이 대부분 축 수준 플롯과  비슷해 보이지만 몇 가지 차이점을 알 수 있습니다. 특히 범례는 플롯 외부에 배치 됩니다. 그들은 또한 모양이 약간 다릅니다.

### Figure 수준 함수가 제공하는 가장 유용한 기능은 여러 서브플롯이 있는 Figure를 쉽게 생성할 수 있는 것입니다. 예를 들어, 동일한 축에 펭귄의 각 종에 대한 세 가지 분포를 쌓는 대신 그림의 열에 각 분포를 플로팅하여 "패싯" 헐수 있습니다.

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```

### Figure 수준 함수는 해당하는 축 수준 함수를 랩핑하고 종류별 키워드 인수 (예: 히스토그램 bin 크기)를 기본 함수로 전달합니다.
### 즉 유연성이 떨어지는 않는다는 장점이 있습니다. 종류별 매개변수가 함수 서명이나 독 스트링에 나타나지 않는 것입니다.
### 일부 기능은 검색하기 어려울 수 있으므로 특정 목표를 달성하는 방법을 이해하기 위해서는 설명서의 서로다른 두페이지를 살펴봐야 할 수 있습니다.
